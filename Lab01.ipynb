# **NumPy Library**

```python
#Task-1
import numpy as np
a = np.random.randint(1, 101, size=20)
b = np.random.randint(1, 101, size=20)
print(a)
print(b)
```

```python
#Task-2
#The reshape function in NumPy allows you to rearrange the dimensions of an array without changing its data.
matrix_a = a.reshape(5, 4)
matrix_b = b.reshape(5, 4)
print(matrix_a)
print(matrix_b)
```

```python
#Task-3
print(matrix_a * matrix_b)
```

```python
#Task-4
modified_matrix_b = matrix_b.T
matrix_c = matrix_a @ modified_matrix_b
print(matrix_c)
```

```python
#Task-5
print(f'Columns 2 and 3 of matrix c:\n{matrix_c[:, 1:3]}')    # Slicing (:, start:end:step) for columns
print(f'Whole matrix c:\n{matrix_c[:, :]}')
```

```python
#Task-6
largest_elem = np.max(matrix_c)   #computes the maximum value in the entire matrix matrix_c
largest_index=np.unravel_index(np.argmax(matrix_c), matrix_c.shape)
#np.argmax(matrix_c) returns the index of the maximum value in the flattened (1D) version of matrix_c
#np.unravel_index(index, matrix_c.shape) converts this flattened index back into the original matrix shape.
#The resulting largest_index is a tuple representing the row and column indices of the largest element in matrix_c.
print(f'Largest Element of matrix_c is: {largest_elem}')
print(f'Index of Largest Element of matrix_c is: {largest_index}')
```

```python
#Task-7
matrix_c_1D = matrix_c.flatten()
print(f'1D array of matrix_c:\n{matrix_c_1D}')
```

# **Matplotlib Library**

```python
#Task-1
import matplotlib.pyplot as plt
temperature_dhaka = [25, 34, 21, 45, 28, 6, 43, 18, 7, 2]
humidity_dhaka = [28, 25,29,20, 26, 50, 19, 29, 52, 55]
plt.scatter(temperature_dhaka, humidity_dhaka, color='r', marker='*')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('Temperature vs Humidity in Dhaka')
plt.show()
```

```python
#Task-2
study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]
plt.figure(figsize=(12, 8))   #This line creates a new figure (plot) with a specified size (12 inches wide and 8 inches tall).
plt.plot(study_hours, marks, marker='o', linestyle='-', color='b')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Study Hours vs Marks')
plt.show()
```

```python
#Task-3
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
subject_scores = [89, 90, 45, 78, 99]
plt.barh(subjects, subject_scores, color='g')   #barh for horizontal bar graph and bar for vertical bar graph
plt.xlabel('Scores')
plt.ylabel('Subjects')
plt.title('Subject Scores')
plt.show()
```

```python
#Task-4
colors = ['red', 'green', 'blue', 'orange', 'purple']
plt.bar(subjects, subject_scores, color=colors)
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.title('Subject Scores')
plt.legend()
plt.show()
```

```python
#Task-5
plt.figure(figsize=(12, 8))
# Scatter plot
plt.subplot(2, 2, 1)
plt.scatter(temperature_dhaka, humidity_dhaka, color='red', marker='*')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('Temperature vs Humidity in Dhaka')
# Line plot
plt.subplot(2, 2, 2)
plt.plot(study_hours, marks, marker='o', linestyle='-', color='b')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Study Hours vs Marks')
# Horizontal bar graph
plt.subplot(2, 2, 3)
plt.barh(subjects, subject_scores, color='g')
plt.xlabel('Scores')
plt.ylabel('Subjects')
plt.title('Subject Scores')
# Vertical bar graph
plt.subplot(2, 2, 4)
plt.bar(subjects, subject_scores, color=colors)
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.title('Subject Scores')
plt.tight_layout()    #This function adjusts the spacing between subplots to prevent overlapping labels
plt.show()
```
